```python
import matplotlib.pyplot as plt
import numpy as np
```

# Euler methoed
## It is derived from the first defination of diffarentiation

$$ \frac{dx}{dt} = f(x,t)$$
Intial condition is at $t = t_0$ the  $x = x_0$
$$x(t+h) = x_0 + hf(x,t)$$

```python
def f(x,t):
    return np.cos(t)
h = 0.001
t_values = np.arange(0,2*np.pi,h)
x_values = []

x = 1 #Intial condition
for t in t_values:
    x_values.append(x)
    x = x + h*f(x,t)

plt.plot(t_values, np.sin(t_values),"green")   
plt.plot(t_values, x_values,"blue")
plt.axhline(c = 'k')
plt.show()
```
